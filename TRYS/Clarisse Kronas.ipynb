```python
import pandas as pd
from subprocess import run
import sys

from biomeanalyzer.metaboanalyst_csv.data_to_csv import (data_to_csv)
from biomeanalyzer.io.load_data import load_legend, load_data
from biomeanalyzer.normalization.copy_normalization import normalize_data
```

```python
novogene_data = load_data(
    '/Users/tiago_silva/Documents/GitHub/BiomeAnalyzer/data/featureTable.sample.total.relative.novogene.xls')
novogene_data
```

```python
novogene_data=novogene_data.reset_index(drop=True)
```

```python
def clean_taxons(string):
    if string == "s__":
        string = "sp."
    elif "__" in string:
        return string.split("__")[1]
    return string
```

```python
split_data = novogene_data['Taxonomy'].str.split(';', expand=True)

# Get the number of resulting columns
num_columns = split_data.shape[1]

# Replace None values with an empty string
split_data = split_data.fillna('')

split_data = split_data.rename(columns={0: '#k', 1: '#p', 2: '#c', 3: '#o', 4: '#f', 5: '#g', 6: '#s'})

split_data = split_data.map(clean_taxons)
```

```python
def clean_taxonomy(taxonomy):
    parts = taxonomy.split(";")

    if parts[-1] == 's__':
        parts[-1] = parts[-2].replace("g__", "") + " sp"

    for i in range(len(parts)):
        if "__" in parts[i]:
            parts[i] = parts[i].split("__")[1]

    return ";".join(parts)

novogene_data['#NAME'] = novogene_data[('Taxonomy')].apply(clean_taxonomy)
novogene_data

novogene_data = novogene_data.set_index("#NAME")
novogene_data = novogene_data.drop(columns=['Taxonomy'])
novogene_data
```

```python
novogene_data = novogene_data * 100
```

```python
normalized_novogene_data = normalize_data(novogene_data)
normalized_novogene_data
```

```python
normalized_novogene_data.reset_index(drop=True, inplace=True)

normalized_novogene_data = pd.concat([normalized_novogene_data, split_data], axis=1)
```
