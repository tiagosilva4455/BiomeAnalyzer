## DADOS EXPERIMENT 2 - NOVOGENE

```python
novogene_data = load_data(
    '/Users/tiago_silva/Documents/GitHub/BiomeAnalyzer/data/featureTable.sample.total.relative.novogene.xls')
novogene_data
```

```python
def clean_taxonomy(taxonomy):
    parts = taxonomy.split(";")
    
    if parts[-1] == 's__':
        parts[-1] = parts[-2].replace("g__", "") + " sp"
    
    for i in range(len(parts)):
        if "__" in parts[i]:
            parts[i] = parts[i].split("__")[1]
 
    return ";".join(parts)
```

```python
novogene_data['Taxonomy'] = novogene_data[('Taxonomy')].apply(clean_taxonomy)
novogene_data
```

```python
novogene_data = novogene_data.rename(columns={"Taxonomy": "#NAME"})
novogene_data
```

```python
novogene_data = novogene_data.set_index("#NAME")
novogene_data
```

```python
novogene_data = novogene_data * 100
novogene_data
```
